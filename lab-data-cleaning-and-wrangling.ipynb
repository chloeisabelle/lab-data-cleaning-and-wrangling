```python
#remove outliers

def remove_outliers(customer_df):
    Q1=np.percentile(customer_df, 25)
    Q3=np.percentile(customer_df,75)
    IQR=Q3-Q1
    customer_df = customer_df[~((customer_df<(Q1-1.5*IQR)) | (customer_df>(Q3+1.5*IQR)))]
    return customer_df
```

```python
customer_df['customer_lifetime_value'] = remove_outliers(customer_df['customer_lifetime_value'])
```
